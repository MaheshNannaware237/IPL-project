```python
import pandas as pd

# Load data
df = pd.read_csv("IPL.csv")

# Basic info and stats
print(df.info())
print(df.describe())
print(df['batting_team'].value_counts())  # Example: Count of matches per team
```

```python
import numpy as np

# Example: Average and Max runs using numpy
runs = df['runs_total'].values
average_runs = np.mean(runs)
max_runs = np.max(runs)

print("Average Runs:", average_runs)
print("Max Runs:", max_runs)

```

```python
import matplotlib.pyplot as plt

# Bar chart: Runs by teams
df.groupby('batting_team')['runs_total'].sum() \
  .sort_values(ascending=False) \
  .plot(kind='bar', figsize=(10, 5))

plt.title('Total Runs by Team')
plt.xlabel('Team')
plt.ylabel('Runs')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Total runs scored by each team
team_runs = df.groupby('batting_team')['runs_total'].sum()

# Plot pie chart
plt.figure(figsize=(10, 10))
plt.pie(team_runs, labels=team_runs.index, autopct='%1.1f%%', startangle=140)
plt.title('Percentage of Total Runs by Each Team')
plt.axis('equal')  # Equal aspect ratio makes it a perfect circle
plt.show()
```
